```python
## with this we can count the Aggregate rating received for each rating
## ex: 110 people have voted as 2.5 ratings and 2148 people have opted not to voted

final_df.groupby(['Aggregate rating']).size().head()
```

```python
## To set an index, we use .reset_index()

final_df.groupby(['Aggregate rating']).size().reset_index().head()
```

```python
## since the column name appears 0, we can rename it with the aggreagate counts for more meaningingful

final_df.groupby(['Aggregate rating']).size().reset_index().rename(columns={0:'Rating Count'}).head()
```

```python
## if we want to see the meaning of aggreagate rating then we will group by the below code:
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings

## since size of groupy is always given when there is numerical values
```

```python
## df_name.sort_values(by column_name, axis=0, ascending=True, inplace=False, 
## kind=’quicksort’, na_position=’last’, ignore_index=False, key=None)

ratings.sort_values(by=['Rating Count'],ascending=False)

## with this we can make out maximum people have voted 3.2 ratings
```

### Observation
- When Rating is between 4.5 to 4.9---> Excellent
- When Rating are between 4.0 to 4.4--->very good
- when Rating is between 3.5 to 3.9----> good
- when Rating is between 2.5 to 3.4----> average
- when Rating is between 1.8 to 2.4----> Poor
- 2148 people have not opted to give rating
- white represents not rated
- red represents poor rated
- orange represents average rated
- yellow represents good rated
- green represents very good rated
- dark green represents excellent rated

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings, hue ='Rating text')
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings, hue ='Rating color', palette =['grey','red','orange','yellow','green','green'])
```

#### Observation:

1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','olive','green'])
```
